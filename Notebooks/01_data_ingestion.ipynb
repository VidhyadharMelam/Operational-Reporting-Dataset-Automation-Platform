# Load Dataset into Spark

```python
# Step 1: Load CSV into Spark DataFrame
df_bronze = spark.read.option("header", True).option("inferSchema", True).csv("/Volumes/workspace/default/volume-1/operational_reporting_raw.csv")

# Step 2: Preview data
df_bronze.show(5)
```

# Save Bronze Table

```python
# Step 4a: Save Bronze table as a managed Delta table
df_bronze.write.format("delta").mode("overwrite").saveAsTable("bronze_operational_reporting")

# Step 4b: Verify table creation
spark.sql("SELECT * FROM bronze_operational_reporting LIMIT 10").show()
```
